# danvas: auto-panel every cell

`canvas.capture_cells()` (alias `danvas.autopanel(canvas)`) registers an
IPython `post_run_cell` hook so **every cell that ends in an expression**
gets its own panel on the canvas — no manual `insert` per cell. Outputs are
rendered with Jupyter's own display machinery, so DataFrames, matplotlib
figures, Plotly figures, and any `_repr_html_` object render as they would
inline.

Panels auto-arrange in a grid. Re-running a cell swaps its panel in place
(keyed on the notebook's stable cell id). Cells ending in a *statement*
(an assignment, a `print`, a loop) produce no value and are skipped.

```python
import danvas

# Start the server in the background, then turn on cell capture.
canvas = danvas.Canvas().serve(port=8000, block=False)
canvas.capture_cells(cols=2, slot_w=400, slot_h=300)   # or: danvas.autopanel(canvas, cols=2)
```

From here on, just write normal cells and watch each output appear as its
own panel in the browser.

```python
# A plain value -> a text Label panel.
sum(range(100))
```

```python
# A rich _repr_html_ object (here a DataFrame) -> an HTML Custom panel.
import pandas as pd
pd.DataFrame({"x": range(5), "x_squared": [i * i for i in range(5)]})
```

```python
# A matplotlib figure -> its PNG, dropped into a Custom panel.
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 4 * np.pi, 400)
fig, ax = plt.subplots()
ax.plot(t, np.sin(t))
ax.set_title("sin(t)")
fig   # end on the figure so it becomes this cell's output
```

```python
# A Plotly figure -> routed through the interactive Plot component.
import plotly.graph_objects as go
go.Figure(data=go.Scatter(y=[2, 1, 3, 5, 4], mode="lines+markers"))
```

## Customising an individual cell

Auto-placement is the default, but any cell can override its own panel with a
`# danvas:` directive line — pin a position/size, lock it, rename it, or opt
out with `skip`. Fields you don't set still fall back to the grid.

```python
# danvas: x=1000 y=920 w=140 h=150 name=metrics label="Pinned metrics" draggable=true
pd.DataFrame({"metric": ["loss", "acc"], "value": [0.12, 0.97]})
```

```python
# danvas: skip
"this value is computed but intentionally NOT mirrored to the canvas"
```

Re-run any cell above — its panel refreshes in place rather than stacking a
duplicate. When you're done, stop capturing (existing panels stay put):
